# **Experiment 07**

## **Build and train a CNN and know the application of different layers**

```python
import tensorflow as tf
import numpy as np
```

### **Step 1: Load the MNIST dataset into your notebook**

```python
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

### **Step 2: Pre-processing and prepare the data for giving to the CNN.**

**a.	Encoding the classes using one hot encoder.**

```python
y_train = tf.keras.utils.to_categorical(y_train, num_classes=10)
y_test = tf.keras.utils.to_categorical(y_test, num_classes=10)
```

**b.	Normalize the features.**

```python
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Reshape the input data to include a channel dimension
x_train = np.expand_dims(x_train, axis=-1)
x_test = np.expand_dims(x_test, axis=-1)
```

### **Step 3: Building the convolutional network model.**

**a.	You may choose the layers.**

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(10, activation='softmax')
])

```

**b.	Print the summary and note the number of neurons and parameters of the model.**

```python
model.summary()
```

**c.	Compile the model and train it using the training data.**

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=10)
```

### **Step 4: Vary the number of layers and repeat step 3.**

```python
model_2 = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),  # Added another convolutional layer
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(10, activation='softmax')
])

model_2.summary()

model_2.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model_2.fit(x_train, y_train, epochs=10)
```

### **Step 5: Implement the architecture of LeNet 5.**

```python
model_lenet = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(6, (5, 5), activation='tanh', input_shape=(28, 28, 1)),
    tf.keras.layers.AveragePooling2D((2, 2)),
    tf.keras.layers.Conv2D(16, (5, 5), activation='tanh'),
    tf.keras.layers.AveragePooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(120, activation='tanh'),
    tf.keras.layers.Dense(84, activation='tanh'),
    tf.keras.layers.Dense(10, activation='softmax')
])

model_lenet.summary()

model_lenet.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model_lenet.fit(x_train, y_train, epochs=10)
```

### **Step 6: Finally note which network gives you the best performance.**

```python
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
print('Model 1 - Test accuracy:', accuracy)

loss_2, accuracy_2 = model_2.evaluate(x_test, y_test, verbose=0)
print('Model 2 - Test accuracy:', accuracy_2)

loss_lenet, accuracy_lenet = model_lenet.evaluate(x_test, y_test, verbose=0)
print('LeNet - Test accuracy:', accuracy_lenet)

# Determine the best performing model
best_accuracy = max(accuracy, accuracy_2, accuracy_lenet)
if best_accuracy == accuracy:
  print("Model 1 has the best performance.")
elif best_accuracy == accuracy_2:
  print("Model 2 has the best performance.")
else:
  print("LeNet has the best performance.")
```

## **Observation and Learning**

Multi-layer perceptron is not accurate for image data due to huge number of inputs. CNNs are better at extracting features like edges and textures. This leads to better performance in tasks like image recognition. <br>
The LeNet-5 CNN architecture has seven layers with tanh as the activtion function.
