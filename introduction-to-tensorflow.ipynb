**Task 3**

```python
def f(x):
  return x**3

def compute_derivative(x):
  x = tf.Variable(x)
  with tf.GradientTape() as tape:
    tape.watch(x)
    y = f(x)
  return tape.gradient(y, x)

def compute_higher_order_derivative(x, order = 2):
  y = x
  for _ in range(order):
    y = compute_derivative(y)
  return y
```

```python
x = 2.0

first_derivative = compute_derivative(x)
print(first_derivative)

second_derivative = compute_higher_order_derivative(x, order = 2)
print(second_derivative)
```
