```python
import tensorflow as tf
```

**Task 1: Building a three-layer feed forward neural network from scratch by initializing random weights and biases.**

```python
input = tf.constant([[0.05, 0.10]])
output = tf.constant([[0.01, 0.99]])
```

```python
w_i_h1 = tf.Variable([[0.15, 0.20],[0.25, 0.30]])
w_h1_o = tf.Variable([[0.40, 0.45],[0.50, 0.55]])

b1 = tf.Variable([0.35, 0.35])
b2 = tf.Variable([0.60, 0.60])
```

```python
h1 = tf.sigmoid(tf.matmul(input, w_i_h1) + b1)

o = tf.sigmoid(tf.matmul(h1, w_h1_o) + b2)
print("Prediction:", o.numpy())
```

**Task 2: Compute the total squared error.**

```python
def total_squared_error(actual_output, expected_output):
    error = actual_output - expected_output
    squared_error = tf.square(error)
    total_squared_error = tf.reduce_sum(squared_error)
    return total_squared_error
```

```python
print("Total Squared Error:", total_squared_error(output, o).numpy())
```

**Task 3: Change the initial weights and biases and compute the error again.**

```python
w_i_h1 = tf.Variable([[0.45, 0.50],[0.85, 0.10]])
w_h1_o = tf.Variable([[0.40, 0.15],[0.50, 0.15]])

b1 = tf.Variable([0.50, 0.70])
b2 = tf.Variable([0.75, 0.55])

h1 = tf.sigmoid(tf.matmul(input, w_i_h1) + b1)

o = tf.sigmoid(tf.matmul(h1, w_h1_o) + b2)
print("Prediction:", o.numpy())

print("Total Squared Error:", total_squared_error(output, o).numpy())
```

**Task 4: Add one more hidden neuron in the middle layer and compare the error.**

```python
w_h1_h2 = tf.Variable(tf.random.normal([2, 2]))
w_h2_o = tf.Variable(tf.random.normal([2, 2]))

b2 = tf.Variable(tf.random.normal([2]))
b3 = tf.Variable(tf.random.normal([2]))
```

```python
h2 = tf.sigmoid(tf.matmul(h1, w_h1_h2) + b2)

o = tf.sigmoid(tf.matmul(h2, w_h2_o) + b3)
print("Prediction:", o.numpy())
```

```python
print("Total Squared Error:", total_squared_error(output, o).numpy())
```

**Task 5: Add one more hidden layer before the output layer and compute the error.**

```python
w_h2_h3 = tf.Variable(tf.random.normal([2, 2]))
w_h3_o = tf.Variable(tf.random.normal([2, 2]))

b3 = tf.Variable(tf.random.normal([2]))
b4 = tf.Variable(tf.random.normal([2]))
```

```python
h3 = tf.sigmoid(tf.matmul(h2, w_h2_h3) + b3)

o = tf.sigmoid(tf.matmul(h3, w_h3_o) + b4)
print("Prediction:", o.numpy())
```

```python
print("Total Squared Error:", total_squared_error(output, o).numpy())
```
